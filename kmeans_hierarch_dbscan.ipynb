# K-Means VS. Hierarchical Cluster VS. DBSCAN

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn import datasets
```

```python
x_random, y_random = datasets.make_moons(n_samples=1500, random_state=0, noise=0.09)
```

```python
px.scatter(x = x_random[:,0], y = x_random[:,1])
```

# K-Means

```python
from sklearn.cluster import KMeans
```

```python
kmeans_random = KMeans(n_clusters=2)
labels = kmeans_random.fit_predict(x_random)
```

```python
px.scatter(x = x_random[:,0], y = x_random[:,1], color=labels, title="K-Means")
```

# Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc_random = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
labels = hc_random.fit_predict(x_random)
```

```python
px.scatter(x = x_random[:,0], y = x_random[:,1], color=labels, title="Hierarchical Clustering")
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan_random = DBSCAN(eps=0.1)
labels = dbscan_random.fit_predict(x_random)
```

```python
np.unique(labels, return_counts=True)
```

```python
px.scatter(x = x_random[:,0], y = x_random[:,1], color=labels, title="DBSCAN")
```

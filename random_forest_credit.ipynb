## Random forest

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.naive_bayes import GaussianNB
import pickle
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
with open('credit.pkl', 'rb') as f:
    x_credit_training, y_credit_training, x_credit_test, y_credit_test = pickle.load(f)
```

```python
x_credit_training.shape, y_credit_training.shape, x_credit_test.shape, y_credit_test.shape
```

```python
random_forest = RandomForestClassifier(n_estimators= 40, 
                                       criterion= 'entropy',
                                       random_state= 0)
random_forest.fit(x_credit_training,y_credit_training)
```

```python
prediction = random_forest.predict(x_credit_test)
prediction
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy_score(y_credit_test,prediction)
```

```python
from yellowbrick.classifier import ConfusionMatrix
confusion_matrix = ConfusionMatrix(random_forest)
confusion_matrix.fit(x_credit_training,y_credit_training)
confusion_matrix.score(x_credit_test,y_credit_test)
```

```python
print(classification_report(y_credit_test,prediction))
```

Note that not always the random forest will perform better than decision tree.

In this case n = 40 accuracy =98.4 vs decision tree accu. =98.2. However with other values of 'n', the accuracy of random forest is smaller than the one from decision tree.
